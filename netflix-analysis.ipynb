```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1.Data Loading

```python
df = pd.read_csv("netflix.csv")

print(df.shape)
df.head()
```

## 2.Data Cleaning

```python
print(df.isnull().sum())
```

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")

df = df.dropna(subset=['date_added', 'rating'])

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
```

## 3.Data Analysis

```python
type_counts = df['type'].value_counts()
year_counts = df['year_added'].value_counts().sort_index()
top_contries = df['country'].value_counts().head(10)

print(type_counts)
```

## 4.Visualization

```python
type_counts.plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.show()
```

```python
year_counts.plot(kind='line')
plt.title("Content Growth Over Time")
plt.show()
```

```python
top_contries.plot(kind='bar')
plt.title("Top Contries")
plt.xticks(rotation = 45)
plt.show()
```

## 5.Insights

```python
print("Insights:")
print("1.Movies are more common than TV Shows.")
print("2.Content increased significantly after 2015.")
print("3.The United States produces the most content.")
```
